Load Pre-Trained Models:

```python
!pip install tensorflow    #ML Package
from tensorflow.keras.applications import ResNet50
from transformers import pipeline

model = ResNet50(weights='imagenet', include_top=False, pooling='avg')
caption_generator = pipeline("image-to-text", model="nlpconnect/vit-gpt2-image-captioning")
```

Image Upload and Display

```python
!pip install opencv-python
import cv2
from google.colab.patches import cv2_imshow

def get_image_path():
  from google.colab import files
  uploaded = files.upload()
  for filename in uploaded.keys():
    image_path = filename
    return image_path
image_path = get_image_path()
image = cv2.imread(image_path)
print(image) #cv2_imshow(image)
```

Caption Generation

```python
!pip install Pillow
from PIL import Image
import numpy as np
from keras.applications.resnet50 import preprocess_input #Import preprocess_input

if image_path:
    image_np = np.array(image) # PIL  to NumPy
    image_np = cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB)  #Convert to RGB
    image = Image.fromarray(image_np)  #Convert back to PIL Image
    image = image.resize((224, 224))
    # image_array = np.array(image) #Convert PIL Image to NumPy array for preprocessing is not needed
    # image_array = preprocess_input(image_array) #Preprocess the image array is not needed
    caption = caption_generator(image)[0]["generated_text"] #Pass PIL Image to caption generator
    print("Caption:", caption)
else:
    print("No image uploaded.")
```
